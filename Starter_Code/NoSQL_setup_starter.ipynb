## Part 2: Update the Database

1. An exciting new halal restaurant just opened in Greenwich, but hasn't been rated yet. The magazine has asked you to include it in your analysis. Add the following restaurant "Penang Flavours" to the database.

```python
# Create a dictionary for the new restaurant data
new_grenenwich_restaurant = {
    "BusinessName":"Penang Flavours",
    "BusinessType":"Restaurant/Cafe/Canteen",
    "BusinessTypeID":"",
    "AddressLine1":"Penang Flavours",
    "AddressLine2":"146A Plumstead Rd",
    "AddressLine3":"London",
    "AddressLine4":"",
    "PostCode":"SE18 7DY",
    "Phone":"",
    "LocalAuthorityCode":"511",
    "LocalAuthorityName":"Greenwich",
    "LocalAuthorityWebSite":"http://www.royalgreenwich.gov.uk",
    "LocalAuthorityEmailAddress":"health@royalgreenwich.gov.uk",
    "scores":{
        "Hygiene":"",
        "Structural":"",
        "ConfidenceInManagement":""
    },
    "SchemeType":"FHRS",
    "geocode":{
        "longitude":"0.08384000",
        "latitude":"51.49014200"
    },
    "RightToReply":"",
    "Distance":4623.9723280747176,
    "NewRatingPending":True
}
```

```python
# Insert the new restaurant into the collection
db.establishments.insert_one(new_grenenwich_restaurant)
```

```python
# Check that the new restaurant was inserted
pprint(establishments.find_one({"AddressLine2":"146A Plumstead Rd"}))
```

2. Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the `BusinessTypeID` and `BusinessType` fields.

```python
# Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the BusinessTypeID and BusinessType fields
for establishment in db.establishments.find({"BusinessType": "Restaurant/Cafe/Canteen"}, {"BusinessType":1, "BusinessTypeID": 1}):
    print(establishment)
```

3. Update the new restaurant with the `BusinessTypeID` you found.

```python
# Update the new restaurant with the correct BusinessTypeID
db.establishments.update_one({'AddressLine2': '146A Plumstead Rd'}, {'$set': {'BusinessTypeID': 1}})
```

```python
# Confirm that the new restaurant was updated
pprint(establishments.find_one({"AddressLine2":"146A Plumstead Rd"}))
```

4. The magazine is not interested in any establishments in Dover, so check how many documents contain the Dover Local Authority. Then, remove any establishments within the Dover Local Authority from the database, and check the number of documents to ensure they were deleted.

```python
# Find how many documents have LocalAuthorityName as "Dover"
query = {'LocalAuthorityName': 'Dover'}
Dover_count = establishments.count_documents(query)
Dover_count
```

```python
# Delete all documents where LocalAuthorityName is "Dover"
establishments.delete_many({"LocalAuthorityName": "Dover"})
```

```python
# Check if any remaining documents include Dover
query = {'LocalAuthorityName': 'Dover'}
Dover_count = establishments.count_documents(query)
Dover_count
```

```python
# Check that other documents remain with 'find_one'
pprint(db.establishments.find_one())
```

5. Some of the number values are stored as strings, when they should be stored as numbers. Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude
db.establishments.update_many({}, [ {'$set':{ "geocode.latitude" : {'$toDecimal': "$geocode.latitude"}}}])

```

```python
# Change the data type from String to Decimal for latitude
db.establishments.update_many({}, [ {'$set':{ "geocode.longitude" : {'$toDecimal': "$geocode.longitude"}}}])
```

```python
# Check that the coordinates are now numbers
pprint(db.establishments.find_one())
```

```python
#List all the distinct values of the field "RatingValue" in the collection "establishments".
db.establishments.distinct( "RatingValue")
```

```python
#Convert all text stored as stings in the "RatingValue" field to null and all numbers stored as strings to integers.
db.establishments.update_many({"RatingValue": {"$regex": ".n$"}}, [ {'$set':{ "RatingValue" : np.nan}}])
db.establishments.update_many({"RatingValue": {"$regex": ".t$"}}, [ {'$set':{ "RatingValue" : np.nan}}])
db.establishments.update_many({"RatingValue": {"$regex": ".s$"}}, [ {'$set':{ "RatingValue" : np.nan}}])
```

```python
# Get a new List of distinct values for the RatingValue field
db.establishments.distinct( "RatingValue")
```

```python
#Change the numeric ratings in the RatingValue field to integers.
db.establishments.update_many({"RatingValue": {"$regex": "[0-6]+"}}, [ {'$set':{ "RatingValue" : {'$toInt': "$RatingValue"}}}])
```

```python
# Check that the RatingValues are now intergers
db.establishments.distinct( "RatingValue")
```
