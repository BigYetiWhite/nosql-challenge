```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
```

### 1. Which establishments have a hygiene score equal to 20?

## Part 3: Exploratory Analysis

```python
# Find the establishments with a hygiene score of 20
hygiene_query = {'scores.Hygiene' : {'$in' : [20]}}
hygiene_results = list(establishments.find(hygiene_query))
pprint(hygiene_results)
```

```python
# Use count_documents to display the number of documents in the result
Hygiene_count = establishments.count_documents(hygiene_query)
Hygiene_count
```

```python
# Display the first document in the results using pprint
pprint(hygiene_results[0])
```

```python
# Convert the result to a Pandas DataFrame
hygiene20_df = pd.DataFrame(hygiene_results)
hygiene20_df
```

```python
# Display the number of rows in the DataFrame

print(f'Number of Records {len(hygiene20_df)}')
```

```python
# Display the first 10 rows of the DataFrame
hygiene20_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
london_query = {'RatingValue' : {'$gte': 4}, 'LocalAuthorityName' : {'$regex' : '.London.'}}
london_rate = list(establishments.find(london_query))
pprint(london_rate)
```

```python
# Use count_documents to display the number of documents in the result
london_count = establishments.count_documents(london_query)
london_count
```

```python
# Display the first document in the results using pprint
pprint(london_rate[0])
```

```python
# Convert the result to a Pandas DataFrame
london4_df = pd.DataFrame(london_rate)
```

```python
# Display the number of rows in the DataFrame
london4_df.info()
```

```python
# Display the first 10 rows of the DataFrame
london4_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of '5', sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Dependencies
degree_search = 0.01
longitude = 0.08384000
latitude = 51.49014200
limit = 5
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

query = {"RatingValue": 5, "latitude": {"$gt": latitude - degree_search, "$lt": latitude + degree_search}, "longitude": {"$gt": longitude - degree_search, "$lt": longitude + degree_search}}
sort =  [("scores.Hygiene", -1)]

Top_5_near_p_flav = list(establishments.find(query).sort(sort).limit(5))
```

```python
# Print the results
pprint(Top_5_near_p_flav)
```

```python
# Convert result to Pandas DataFrame
Top5_df = pd.DataFrame(Top_5_near_p_flav)
Top5_df
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
match_query = {'$match': {'scores.hygiene': 0}}
# 2. Groups the matches by Local Authority
group_query = {'$group': {'_id': '$local_authority_name', 'count': {'$sum': 1}}}
# 3. Sorts the matches from highest to lowest
sort_query = {'$sort': {'count': -1}}
# Print the number of documents in the result
limit_query = {'$limit': 10}
pipeline = [match_query, group_query, sort_query, limit_query]
LA_results = list(establishments.aggregate(pipeline))
```

```python
# Print the first 10 results
pprint(LA_results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
LA_df = pd.DataFrame(list(LA_results))
# Display the number of rows in the DataFrame
print("Number of rows in result: ", len(LA_df))
# Display the first 10 rows of the DataFrame
LA_df.head(10)
```
